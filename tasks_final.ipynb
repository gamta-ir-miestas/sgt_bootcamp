1) Use map to double each element in a list of numbers.
Use filter to select only the odd numbers from a list of integers.

```python
import numpy as np
number_list = list(np.random.randint(11, size = 20))

def double(num):
    return num * 2

result = list(map(double, number_list))
print(result)

filtered = list(filter(lambda num: num %2 != 0, number_list))
print(filtered)
```

5)Use a deque to implement a queue and perform enqueue and dequeue operations. Reverse the order of elements in a deque.

```python
from collections import deque

queue = deque()

queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)

print("Queue:", queue)

item = queue.popleft()
print("Dequeued:", item)

item = queue.popleft()
print("Dequeued:", item)

print("Queue after dequeue:", queue)

queue.reverse()
print("Reversed Queue:", queue)
```

6)Create a defaultdict that maps a character to a list of words starting with that character. Calculate the sum of values for each key in a dictionary using a defaultdict.


```python
from collections import defaultdict

words = ["apple", "zebra", "cupid", "school", "angel", "banana", "cat", "dog", "elephant", "frog"]

word_dict = defaultdict(list)

for word in words:
    word_dict[word[0]].append(word)

for key, value in word_dict.items():
    print(f"Words starting with '{key}': {value}")
    print(f"Total number of words starting with '{key}': {len(value)}")

```

7)Use reduce to find the maximum element in a list of numbers. Combine a list of strings into a single string using reduce.

```python
from functools import reduce

def bigger(a, b):
    return a if a > b else b

numbers = [123, 54, 678, 999, 1, 15, 1001, 67]

maximum = reduce(bigger, numbers)
print(maximum)
```

```python
strings = ['bgdfdhd', 'cgweuyegcu', 'sksgwwrtwqn', 'dsbhsafdqyte']

def add(a, b):
    return a+b
    
summ = reduce(add, strings)
print(summ)
```

```python
condition = list(df['conditionCode'])
conset = list(set(condition))
condition_df = pd.DataFrame(conset)
condition_df.columns = ['conditionCode']
condition_df = condition_df.rename_axis('id')
condition_df = condition_df.reset_index()
```

```python
from sqlalchemy import create_engine
from sqlalchemy import text

engine = create_engine("mysql+pymysql://root:@localhost/bootcamp")
condition_df.to_sql(name="weather_conditions",con=engine,if_exists="append",index = False)
```
